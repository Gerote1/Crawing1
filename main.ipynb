```python
import plotly.express as px 
import pandas as pd 

# Load dataset
df = pd.read_csv("results.csv")

df['Domain'] = df['Domain'].str.replace(r'^www\.', '', regex=True)

# Bar chart for domain frequency
domain_count = df['Domain'].value_counts().reset_index()



df['Domain'] = df['Domain'].str.replace(r'^www\.', '', regex=True)
# Rename columns
domain_count.columns = ['Domain', 'Frequency']

# Select the top 50 domains
top_50_domains = domain_count.nlargest(200, 'Frequency')

# Create the bar chart for the top 50 domains
fig = px.bar(top_50_domains, x='Domain', y='Frequency', title='Top 50 Domain Frequency',labels={'Domain': 'Domain', 'Frequency': 'Frequency'})
fig.show()
```

```python
## word cloud for title and description
##Visualize the most frequent keywords in the Title and Description columns to understand common topics and keyword opportunities.
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud for Title
title_text = " ".join(df['Title'].fillna(''))
wordcloud_title = WordCloud(width=400, height=400, background_color='white').generate(title_text)

# Generate word cloud for Description
description_text = " ".join(df['Description'].fillna(''))
wordcloud_description = WordCloud(width=400, height=400, background_color='white').generate(description_text)

# Plot word clouds side by side
plt.figure(figsize=(15, 7))

# Word cloud for Title
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_title, interpolation='bilinear')
plt.title('Title Word Cloud')
plt.axis('off')

# Word cloud for Description
plt.subplot(1, 2, 2)
plt.imshow(wordcloud_description, interpolation='bilinear')
plt.title('Description Word Cloud')
plt.axis('off')

plt.tight_layout()
plt.show()

```
